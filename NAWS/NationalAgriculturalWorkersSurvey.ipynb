### Import necessary modules

```python
import pandas as pd
```

### Read in data and merge dataframes

```python
# read in data
# specify low_memory=False because guessing dtype is memory intensive
A2E = pd.read_csv("NAWS_A2E1103.csv", low_memory=False)
F2Y = pd.read_csv("NAWS_F2Y1103.csv", low_memory=False)

# change the ID variable to a string for merging
A2E['FWID'] = A2E['FWID'].astype(str)
F2Y['FWID'] = F2Y['FWID'].astype(str)

# print the length of each dataframe
print("The length of A2E is", len(A2E))
print("The length of F2Y is", len(F2Y))

# merge the two dataframes
merged_df = pd.merge(A2E, F2Y, on='FWID', how='inner')

# print the length of the merged dataframe
print("The length of the merged_df is", len(merged_df))

# display the merged df
pd.set_option('display.max_columns', None) # set to show all columns
merged_df
```

### Save the merged dataframe as pickle

```python
merged_df.to_pickle('NAWS_df.pkl')
```

### Read back in pickle dataframe

```python
merged_df = pd.read_pickle('NAWS_df.pkl')

# display the merged df from pickle
pd.set_option('display.max_columns', None) # set to show all columns
merged_df
```
